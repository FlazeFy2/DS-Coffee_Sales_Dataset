```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Coffe_sales.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['cash_type','coffee_name','Time_of_Day','Weekday']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Comparison Of {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Sales Distribution : Weekend vs Weekday
df = pd.read_csv("Coffe_sales.csv")

# Categorize Day Weekday or Weekend
def categorize_day(val):
    if val == "Sun" or val == "Sat":
        return "Weekend"
    else:
        return "Weekday"

df["Day Category"] = df["Weekday"].apply(categorize_day)

# Count the number of sales
sales_counts = df["Day Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
sales_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution : Weekend vs Weekday")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Sales Distribution : Time Of Order
df = pd.read_csv("Coffe_sales.csv")

# Categorize Time Of Order
def categorize_day(val):
    if val <= 12:
        return "Morning"
    elif val > 12 and val <= 18:
        return "Afternoon"
    else:
        return "Evening"

df["Time Category"] = df["hour_of_day"].apply(categorize_day)

# Count the number of sales
sales_counts = df["Time Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
sales_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution : Time Of Order")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Total Revenue Between Cafeinn and Non-Cafeinn Drink
df = pd.read_csv("Coffe_sales.csv")

# Categorize Coffee vs Non-Coffee
coffee_items = [
    "Espresso", "Americano", "Americano with Milk", "Latte", "Cappuccino", "Mocha", "Cortado", "Macchiato", "Flat White", "Long Black"
]

# Categorization
def categorize_coffee(name):
    if name in coffee_items:
        return "Cafeinn"
    else:
        return "Non-Cafeinn"

df["Cafeinn Category"] = df["coffee_name"].apply(categorize_coffee)

# Custom autopct function to show both percentage and nominal
def autopct_format(values):
    def inner(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n(${val:,})'
    return inner

# Pie Chart
plt.figure(figsize=(6, 6))
sales_sum = df.groupby("Cafeinn Category")["money"].sum().sort_index()
sales_sum.plot.pie(
    autopct=autopct_format(sales_sum),
    startangle=90,
    textprops={'fontsize': 11}
)
plt.title("Comparison of Total Revenue Between Cafeinn and Non-Cafeinn Drink")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Total Revenue Between Quartal All Year
df = pd.read_csv("Coffe_sales.csv")

# Categorize Quartal
def categorize_day(val):
    if val <= 3:
        return "Q1"
    elif val <= 6:
        return "Q2"
    elif val <= 9:
        return "Q3"
    else:
        return "Q4"

df["Quartal"] = df["Monthsort"].apply(categorize_day)

# Custom autopct function to show both percentage and nominal
def autopct_format(values):
    def inner(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n(${val:,})'
    return inner

# Pie Chart
plt.figure(figsize=(6, 6))
sales_sum = df.groupby("Quartal")["money"].sum().sort_index()
sales_sum.plot.pie(
    autopct=autopct_format(sales_sum),
    startangle=90,
    textprops={'fontsize': 11}
)
plt.title("Comparison of Total Revenue Between Quartal All Year")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
