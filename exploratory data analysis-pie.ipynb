```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Coffe_sales.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['cash_type','coffee_name','Time_of_Day','Weekday']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Comparison Of {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Sales Distribution : Weekend vs Weekday
df = pd.read_csv("Coffe_sales.csv")

# Categorize Day Weekday or Weekend
def categorize_day(val):
    if val == "Sun" or val == "Sat":
        return "Weekend"
    else:
        return "Weekday"

df["Day Category"] = df["Weekday"].apply(categorize_day)

# Count the number of sales
sales_counts = df["Day Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
sales_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution : Weekend vs Weekday")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Total Revenue Between Cafeinn and Non-Cafeinn Drink
df = pd.read_csv("Coffe_sales.csv")

# Categorize Coffee vs Non-Coffee
coffee_items = [
    "Espresso", "Americano", "Americano with Milk", "Latte", "Cappuccino", "Mocha", "Cortado", "Macchiato", "Flat White", "Long Black"
]

# Categorization
def categorize_coffee(name):
    if name in coffee_items:
        return "Cafeinn"
    else:
        return "Non-Cafeinn"

df["Cafeinn Category"] = df["coffee_name"].apply(categorize_coffee)

# Custom autopct function to show both percentage and nominal
def autopct_format(values):
    def inner(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n(${val:,})'
    return inner

# Pie Chart
plt.figure(figsize=(6, 6))
sales_sum = df.groupby("Cafeinn Category")["money"].sum().sort_index()
sales_sum.plot.pie(
    autopct=autopct_format(sales_sum),
    startangle=90,
    textprops={'fontsize': 11}
)
plt.title("Comparison of Total Revenue Between Cafeinn and Non-Cafeinn Drink")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# EDA - Pie Chart Comparison of Total Revenue Rush Hour vs Non-Rush Hour (Weekday vs Weekend)
df = pd.read_csv("Coffe_sales.csv")

# Categorize Rush Hour
def categorize_rush(hour):
    if (7 <= hour <= 9) or (12 <= hour <= 13) or (17 <= hour <= 19):
        return "Rush Hour"
    else:
        return "Non-Rush Hour"

df["Rush Category"] = df["hour_of_day"].apply(categorize_rush)

# Categorize Weekday / Weekend
def categorize_week(day):
    if day in ["Sat", "Sun"]:
        return "Weekend"
    else:
        return "Weekday"

df["Day Type"] = df["Weekday"].apply(categorize_week)

# Combine Category
df["Time Category"] = df["Rush Category"] + " (" + df["Day Type"] + ")"

# Custom autopct function to show both percentage and nominal
def autopct_format(values):
    def inner(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n(${val:,})'
    return inner

# Pie Chart
plt.figure(figsize=(7, 7))
sales_sum = df.groupby("Time Category")["money"].sum().sort_index()
sales_sum.plot.pie(
    autopct=autopct_format(sales_sum),
    startangle=90,
    textprops={'fontsize': 11}
)

plt.title("Comparison of Total Revenue\nRush Hour vs Non-Rush Hour (Weekday vs Weekend)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
