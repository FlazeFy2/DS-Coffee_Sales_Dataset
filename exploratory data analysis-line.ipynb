```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Line Chart Sales Distribution per Hour by Coffee Name
df = pd.read_csv("Coffe_sales.csv")

# Count Sales per Coffee Name
time_counts = df.groupby(["hour_of_day", "coffee_name"], observed=False).size().unstack(fill_value=0)

# Sort by hour
time_counts = time_counts.sort_index()

# Line Chart
plt.figure(figsize=(12, 6))
for coffee_name in time_counts.columns:
    plt.plot(time_counts.index, time_counts[coffee_name], marker='o', label=coffee_name)
# Add labels
for coffee_name in time_counts.columns:
    for x, value in zip(time_counts.index, time_counts[coffee_name]):
        plt.text(x, value + 0.5, str(value), ha='center', va='bottom', fontsize=8)
plt.title("Sales Distribution per Hour by Coffee Name")
plt.xlabel("Time (Hour of Day)")
plt.ylabel("Number of Sales")
plt.xticks(time_counts.index, rotation=45)
plt.legend(title="Coffee Name")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Line Chart Sales Distribution per Hour by Day Type
df = pd.read_csv("Coffe_sales.csv")

# Categorize Weekday or Weekend
def categorize_day(val):
    if val == "Sat" or val == "Sun":
        return "Weekend"
    else:
        return "Weekday"

df["Day Type"] = df["Weekday"].apply(categorize_day)

# Count Sales per Coffee Name
time_counts = df.groupby(["hour_of_day", "Day Type"], observed=False).size().unstack(fill_value=0)

# Sort by hour
time_counts = time_counts.sort_index()

# Line Chart
plt.figure(figsize=(12, 6))
for coffee_name in time_counts.columns:
    plt.plot(time_counts.index, time_counts[coffee_name], marker='o', label=coffee_name)
# Add labels
for coffee_name in time_counts.columns:
    for x, value in zip(time_counts.index, time_counts[coffee_name]):
        plt.text(x, value + 0.5, str(value), ha='center', va='bottom', fontsize=8)
plt.title("Sales Distribution per Hour by Day Type")
plt.xlabel("Time (Hour of Day)")
plt.ylabel("Number of Sales")
plt.xticks(time_counts.index, rotation=45)
plt.legend(title="Day Type")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Line Chart Sales Distribution per Hour by Cafeinn Category
df = pd.read_csv("Coffe_sales.csv")

# Categorize Coffee vs Non-Coffee
coffee_items = [
    "Espresso", "Americano", "Americano with Milk", "Latte", "Cappuccino", "Mocha", "Cortado", "Macchiato", "Flat White", "Long Black"
]

# Categorization
def categorize_coffee(name):
    if name in coffee_items:
        return "Cafeinn"
    else:
        return "Non-Cafeinn"

df["Cafeinn Category"] = df["coffee_name"].apply(categorize_coffee)

# Count Sales per Coffee Name
time_counts = df.groupby(["hour_of_day", "Cafeinn Category"], observed=False).size().unstack(fill_value=0)

# Sort by hour
time_counts = time_counts.sort_index()

# Line Chart
plt.figure(figsize=(12, 6))
for coffee_name in time_counts.columns:
    plt.plot(time_counts.index, time_counts[coffee_name], marker='o', label=coffee_name)
# Add labels
for coffee_name in time_counts.columns:
    for x, value in zip(time_counts.index, time_counts[coffee_name]):
        plt.text(x, value + 0.5, str(value), ha='center', va='bottom', fontsize=8)
plt.title("Sales Distribution per Hour by Cafeinn Category")
plt.xlabel("Time (Hour of Day)")
plt.ylabel("Number of Sales")
plt.xticks(time_counts.index, rotation=45)
plt.legend(title="Cafeinn Category")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Line Chart Sales Distribution per Hour by Semester
df = pd.read_csv("Coffe_sales.csv")

# Categorization
def categorize_day(val):
    if val <= 6:
        return "S1"
    else:
        return "S2"

df["Semester"] = df["Monthsort"].apply(categorize_day)

# Count Sales per Coffee Name
time_counts = df.groupby(["hour_of_day", "Semester"], observed=False).size().unstack(fill_value=0)

# Sort by hour
time_counts = time_counts.sort_index()

# Line Chart
plt.figure(figsize=(12, 6))
for coffee_name in time_counts.columns:
    plt.plot(time_counts.index, time_counts[coffee_name], marker='o', label=coffee_name)
# Add labels
for coffee_name in time_counts.columns:
    for x, value in zip(time_counts.index, time_counts[coffee_name]):
        plt.text(x, value + 0.5, str(value), ha='center', va='bottom', fontsize=8)
plt.title("Sales Distribution per Hour by Semester")
plt.xlabel("Time (Hour of Day)")
plt.ylabel("Number of Sales")
plt.xticks(time_counts.index, rotation=45)
plt.legend(title="Semester")
plt.tight_layout()
plt.show()
```
